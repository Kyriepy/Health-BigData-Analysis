## 1.if 제어문

### 1.1 일반제어문

```python
myscore = input('성적은 ')
myscore = int(myscore)

if (myscore >= 80):
    print( "잘했어요")
else:
    print("다음엔 더 열심히")
```

### 1.2 다중제어문

```python
if (myscore >= 90):
    print('A')
elif (myscore >= 80):
    print('B')
else:
    print('C')
```

## 2.for 반복문

### 2.1 기본 반복문

```python
cal=0

for i in range(101):
    cal += i

print(cal)
```

### 2.2 리스트 이용

## 3.함수와 클래스

### 3.1 반환값이 없는 함수

```python
def avg(x, y):
    print((x + y)/2)
    
avg(4, 5)
```

### 3.2 반환값이 있는 함수

```python
def avg1(x, y):
    return((x + y)/2)

a = avg1(7,5)
print(a)
```

### 3.3 객체지향 클래스

```python
class Cal():

    def __init__(self, x1, x2): # init: 초기값 세팅, self=자기객체를 의미
        self.x1 = x1
        self.x2 = x2
        
    def add(self):
        result = self.x1 + self.x2
        return result
    
    def sub(self):
        result = self.x1 - self.x2
        return result
```

```python
# 객체 생성
cal = Cal(3,5)

print(cal.add())
print(cal.sub())
```

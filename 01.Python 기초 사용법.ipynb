# 3.데이터 포멧팅

```python
print('이름은 ' + name +"입니다")
```

```python
# 문자형 형태로 변환 필요
print('tot1: ' + tot1)
```

```python
print('tot1의 타입은' + type(tot1))
```

```python
print('tot1: ' + str(tot1))
print('tot1의 타입은' + str(type(tot1)))
```

```python
# 기본 출력
# ,로 연결하면 변환없이 출력 가능
print('tot1:', tot1) # 1칸 띄어쓰기 처리
print('tot1의 타입은', type(tot1), ' 입니다.')
```

```python
# 문자열 포멧 이용
print('tot1: %d' % tot1)
print('tot1: %d' % tot2)
print('tot1: %0.2f' % tot2)
print('이름: %s' % name)
print('tot1의 값: %0.2f, 이름: %s' % (tot1,name) )
```

```python
# format 이용
print('tot1의 값: {0}'.format(tot1))
print('tot2의 값: {0:0.2f}, 이름: {1}'.format(tot2, name))

#.format 대산에 f 이용
print('tot2의 값: {0.2f}, 이름: {name}')
print(f'tot2의 값: {tot2:0.2f}, 이름: {name}')
```

```python
print('tot1의 값: {0}'.format(tot1))

tot2 = [1,2,3,4]

print('tot2의 값: {0}')
```

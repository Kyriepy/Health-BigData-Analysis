```python
import pandas as pd

weather_df = pd.read_csv("./data/00_2.weather.csv")

weather_df.head(10)
```

```python
weather_df.info()
```

```python
filter = (weather_df.origin == 'EWR') | (weather_df.origin == 'JFK')

weather_df.loc[filter, ['origin', 'year', 'month', 'day', 'hour', 'temp']]
```

```python
weather_df['temp'] = (weather_df['temp'] - 32) / 1.8
weather_df
```

```python
weather_df.groupby('month')['temp'].mean().sort_values(ascending= False)
```

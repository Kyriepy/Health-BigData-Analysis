```python
tips = sns.load_dataset("tips")
tips
```

```python
# basic Graph
# 1. kind = scatter

sns.set_theme(style="whitegrid")
sns.scatterplot(x = "total_bill",
                y = "tip",
                data = tips)

plt.show()
```

```python
# 2. relplot + kind

sns.set_theme(style="darkgrid")
sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            kind = "scatter")


plt.show()
```

```python
# 3. 

g = sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            kind = "scatter")

g.set(title = "food",
      xlabel = "food ",
      ylabel = "tip")
```

```python
sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            color = "green",
            alpha = .3)

plt.show()
```

```python
sns.relplot(x = "total_bill",
            y = "tip",
            hue = "day",
            data = tips)

plt.show()
```

```python
sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            hue = "day",
            style= "time",
            size = "size")
```

## window slice

```python
sns.relplot(x = "total_bill",
            y = "tip",
            data = tips,
            col = "smoker")

plt.show()
```

```python
g = sns.FacetGrid(col = "day",
                  col_wrap=3,
                  hue = "smoker",
                  height = 3,
                  aspect = 0.8,
                  data = tips)

g.map(sns.scatterplot,
      "total_bill",
      "tip",
      alpha = .7)

g.add_legend()
plt.show()
```

```python
sns.displot(x = "total_bill",
            kind = "hist",
            data = tips,
            binwidth= 10)

plt.show()
```

```python
sns.displot(x = "total_bill",
            kind = "hist",
            data = tips,
            bins = 10,
            hue = "sex")
```

```python
sns.displot(x = "total_bill",
            kind = "hist",
            data = tips,
            bins = 10,
            hue = "sex",
            col = "sex")
```

```python
# 4.2 kernal density estimation 

sns.displot(x = "total_bill",
            kind = "kde",
            data = tips)
```

```python
sns.displot(x = "total_bill",
            fill = True,
            kde = True,
            kind = "hist",
            data = tips)
```

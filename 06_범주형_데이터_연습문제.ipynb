# 범주형 자료 분석

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme(style="darkgrid")
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # minus 인식
```

```python
# 데이터 프레임으로 저장

og_df = pd.read_csv("./data/01_2.OnlineGame.csv", encoding = 'cp949')
og_df.head(10)
```

## 1. 성별 : 1 = 남자 , 2 = 여자

```python
og_df['성별'].replace({1:'남자', 2:'여자'}, inplace = True)

og_df.head()
```

## 2. 학력 : 1 = 고등학생 이하, 2 = 고등학교 졸업 , 3 = 대학생, 4 = 대학생 이상

```python
og_df['학력'].replace({1:'고등학생 이하', 2:'고등학교 졸업', 3:'대학생', 4:'대학생 이상'}, inplace = True)
og_df.head()
```

## 3. 성별의 돗수분포표, 막대그래프(돗수), 막대그래프(비율)

```python
freq_table = pd.DataFrame(og_df['성별'].value_counts())
freq_table.columns = ['count']
freq_table
```

```python
sns.catplot(x = '성별',
            kind = 'count',
            data = og_df)
plt.show()
```

```python
freq_table['prop'] = np.round(freq_table['count']/sum(freq_table['count']), 2)
freq_table
```

```python
sns.catplot(x = '성별',
            y = 'prop',
            kind = 'bar',
            data = freq_table)
plt.show()
```

## 4. 학력의 돗수분포표, 막대그래프(돗수), 막대그래프(비율)

```python
freq_table1 = pd.DataFrame(og_df['학력'].value_counts())
freq_table1.columns = ['count']
freq_table1
```

```python
sns.catplot(x = '학력',
            kind = 'count',
            data = og_df)
plt.show()
```

```python
freq_table1['prop'] = np.round(freq_table1['count']/sum(freq_table1['count']), 2)
freq_table1
```

```python
sns.catplot(x = '학력',
            y = 'prop',
            kind = 'bar',
            data = freq_table1)
plt.show()
```

```python
# 두 변수 모두 범주형일 때
cross_table = pd.crosstab(index = og_df['성별'],
                          columns = og_df['학력'],
                          margins = True)
cross_table
```

```python
# 정리
cross_table.index = ['남성', '여성', '열 전체']
cross_table.columns = ['고등학교 졸업', '고등학생 이하','대학생', '대학생 이상', '행 전체']
cross_table
```

```python
cross_prop = cross_table/cross_table.loc["열 전체", "행 전체"]
cross_prop
```

```python
sns.catplot(x = '성별',
            hue = '학력',
            kind = 'count',
            data = og_df)
plt.show()
```

# 05. 그래프 그리기 (matplotlib) 

## 1. 피키지 설치 및 데이터 불러오기

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 에러화면 세팅

import warnings

warnings.simplefilter(action = 'ignore', category = FutureWarning)
warnings.simplefilter(action = 'ignore', category = UserWarning)

#테마설정
sns.set_theme(style = 'darkgrid')

#한글 인식
plt.rc('font', family = 'Malgun Gothic')
```

```python
tips = sns.load_dataset("tips")
```

## 2. 기본 그래프 그리기

```python
# 기본형
plt.scatter(x = "total_bill",
            y = "tip",
            data = tips)
```

```python
# 객체 관점 스타일 : x , y 제거, 스타일 지정 가능

fig, ax = plt.subplots()

ax.scatter("total_bill",
           "tip",
           data = tips)
```

```python
# list

x = tips.total_bill.tolist()
y = tips.tip.tolist()

fig, ax = plt.subplots()
ax.scatter(x,
           y)

plt.show()
```

```python
data = {'total_bill':tips.total_bill,
        'tip': tips.tip}

fig, ax = plt.subplots()
ax.scatter("total_bill",
           "tip",
           data = data)

ax.set(title = "음식값과 팁의 산점도 그래프",
       xlabel = "음식값",
       ylabel = "팁",
       xlim = (0,50),
       ylim = (0,10))
```

```python
data = {'total_bill':tips.total_bill,
        'tip': tips.tip}

fig, ax = plt.subplots()
ax.scatter("total_bill",
           "tip",
           data = data,
           c = "red", s = 20, marker = "x")

ax.set(title = "음식값과 팁의 산점도 그래프",
       xlabel = "음식값",
       ylabel = "팁",
       xlim = (0,50),
       ylim = (0,10))
```

```python
x1 = tips[tips.sex == "Male"].total_bill
x2 = tips[tips.sex == "Female"].total_bill
y1 = tips[tips.sex == "Male"].tip
y2 = tips[tips.sex == "Female"].tip

fig, ax = plt.subplots(figsize = (8,6))
ax.scatter(x1, y1,
           c = "b", alpha=0.5, marker='o', label = "Male")
ax.scatter(x2, y2,
           c = "r", alpha=0.5, marker="s", label = "Female")
ax.legend()
```

```python
# 그래프 분리 (열 분리)

fig, ax = plt.subplots(1,2,figsize = (8,6), sharey = True)
fig.suptitle('plot of total_bill and tip')

ax[0].scatter(x1, y1,
              c = "b", alpha = 0.5, marker = "o", label = "Male")

ax[0].set(title = "Male",
          xlabel = 'total_bill',
          ylabel = 'tip',
          xlim = (0,50),
          ylim = (0,10))

ax[1].scatter(x2, y2,
              c = "r", alpha = 0.5, marker = "o", label = "Female")

ax[1].set(title = "Female",
          xlabel = 'total_bill',
          ylabel = 'tip',
          xlim = (0,50),
          ylim = (0,10))

fig.legend()
```

```python
# 그래프 분리 (행 분리)

fig, (ax1, ax2) = plt.subplots(2,figsize = (8,6), sharey = True)
fig.suptitle('plot of total_bill and tip')

ax1.scatter(x1, y1,
              c = "b", alpha = 0.5, marker = "o", label = "Male")

ax1.set(title = "Male",
          xlabel = 'total_bill',
          ylabel = 'tip',
          xlim = (0,50),
          ylim = (0,10))

ax2.scatter(x2, y2,
              c = "r", alpha = 0.5, marker = "o", label = "Female")

ax2.set(title = "Female",
          xlabel = 'total_bill',
          ylabel = 'tip',
          xlim = (0,50),
          ylim = (0,10))
```

## 4. 다양한 함수 그래기

```python
cat = tips['day'].value_counts()

labels = cat.index

fig, ax = plt.subplots(figsize = (8,6))
bar_colors = ['red', 'blue', 'green', 'orange']
bar_container = ax.bar(labels, cat,
                       color = bar_colors, alpha = 0.5)

ax.set(title = 'barplot')
ax.bar_label(bar_container, fmt = '{:,.0f}')
```

```python
labels = cat.index
fig, ax = plt.subplots(figsize = (8,6))
ax.pie(cat, labels = labels)
ax.set(title = 'pieplot')
ax.legend()
```
